```python
# Loading the given csv file using pandas 'read_csv()' function and displaying first 10 rows of the data.
# Pandas gave an error while reading the CSV, because by default it assumes that the encoding is UTF-8. 
# To decode it and read it properly additional parameter is given. 
df=pd.read_csv("turnout.csv", encoding = 'ISO-8859-1')
df.head(10)
```

### D.Find mean, median and mode for atleast one possible column for each (3 marks)

### Check whether the dataset having null values or not. If yes, do replace them with suitable average value (2 marks) (*Think out of box)

```python
df.isnull().sum()
```

```python
#Filling the nulls with mean of corresponding column as they are numerical columns and no outliers are observed.
#Defined a generic function to fill nulls to avoid repetitive coding. 

def fillnulls(df_fun,col):
    mean_value=df_fun[col].mean() 
    df_fun[col].fillna(mean_value,inplace=True)
    
fillnulls(df,'extraversion')
fillnulls(df,'independ')
fillnulls(df,'selfcontrol')
fillnulls(df,'anxiety')
fillnulls(df,'novator')
```

```python
df.isnull().sum()
```

### E.Check the outliers and remove it from the dataset. (3 marks)

```python
# For the data analysis visualizing the potential outliers first going for box plot
df_numeric=df.select_dtypes(exclude=object)
df2_numeric = df_numeric.drop(["Employee ID","event","greywage"],axis=1)
df2_numeric
sns.boxplot(df2_numeric)

# From the box plot we see only age is having outliers, i.e above ~ 50 and below ~8
# Remaining all numeric variables are under IQR limits.
```

```python
# Removing the outliers for age. 
q1=df["age"].quantile(0.25)
q3=df["age"].quantile(0.75)

# Finding Inter quartile region
iqr = q3-q1

# Using the standards identifying the upper and lower limits. 

u_limit= q3 + 1.5*iqr
l_limit= q1 - 1.5*iqr
print("Upper limit above which the Outliers to be removed is:" , u_limit)
print("Upper limit above which the Outliers to be removed is:" , l_limit)

df_filtered = df[df['age'].between(l_limit,u_limit)]
print("Final size(shape) of the Dataframe after doing needful dataclean up is: ", df_filtered.shape)

df_filtered
```

### F. A city affected by 3 cyclones in a year on an average. Find the probability if the same city will getting affected by exactly 2 cyclones in the upcoming year (4 marks)

```python
# As the number of possibilities are unknown applying poisson distribution theory  
import scipy.stats as stats
cyclone_avg = 3
cyclone_prob = 2
p1 = stats.poisson.pmf(cyclone_avg,cyclone_prob)
print(p1)
```

### G.If an athlete attended 6 olympics in his lifetime. Find the probability if he exactly have 2 wins (4 marks)

```python
# As athlete has probability of either winning or loosing (bi) Applying Binomial Distribution 
attended=6
prob=0.5 #(1/2 win or loose, this is assumed as it not available from the question)
desired_wins=2
wins_probability = stats.binom.pmf(desired_wins,attended,prob)
print(wins_probability)
```
