```python
df2.sort_values('Price', ascending=False)
```

```python
df2.sort_values(['Category','Price'], ascending=[True, False])
```

```python
sns.barplot(data=df2, x='Category', y='Price')
```

```python
x=sns.barplot(data=df2, x='Category', y='Price')
for i in x.containers:
    plt.bar_label(i)
plt.show()
```
