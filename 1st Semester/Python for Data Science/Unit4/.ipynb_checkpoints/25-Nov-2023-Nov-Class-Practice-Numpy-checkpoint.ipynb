1. Write a function that takes start and end of a range returns a Pandas series object

containing numbers within that range.

In case the user does not pass start or end or both they should default to 1 and 10 respectively.

eg.

range_series() -> Should Return a pandas series from 1 to 10

range_series(5) -> Should Return a pandas series from 5 to 10

range_series(5, 10) -> Should Return a pandas series from 5 to 10.

```python
def range_series(st=1, en=10):
    return(pd.Series(np.arange(st,en)))
    
```

```python
range_series()
```

```python
range_series(5)
```

```python
range_series(5,10)
```

2. Create a function that takes in two lists named keys and values as arguments.

Keys would be strings and contain n string values.

Values would be a list containing n lists.

The methods should return a new pandas dataframe with keys as column names and values as their corresponding values

e.g. -> create_dataframe(["One", "Two"], [["X", "Y"], ["A", "B"]]) -> should return a dataframe

One Two

0 X Y

1 A B

```python
def create_dataframe(keys, values):  
    df=pd.DataFrame(values,columns=keys)
    df.reset_index()
    return(df)
```

```python
df1=create_dataframe(["One", "Two"], [["X", "Y"], ["A", "B"]])
```

```python
df1
```

5. Write a method that will take a column name as argument and return the name of the column with which the
given column has the highest correlation.

The data to be used is the cars dataset.

The returned value should not the column named that was passed as the parameters.

E.G: get_max_correlated_column('mpg') -> should return 'drat'

```python
corr_df= df.corr()
```

```python
corr_df
```

```python
def get_max_correlated_column(column):
    #Converting the column to dataframe again as the passed on column will come as Series object
    local_df = pd.DataFrame(column)
    print(local_df, '')

    #Deleting the row which contains the self named rows as the self correlation is to be moved. 
    given_column_name=local_df.columns[0]
    updated_local_df=local_df.drop(labels=[given_column_name])

    #Getting the row with max correction location
    max_corr_location= updated_local_df.loc[updated_local_df[given_column_name] == updated_local_df[given_column_name].max()]
    return(max_corr_location)
    
    #return(max_corr_location.index[0]) ## Use this if you just need the variable name which is highly correlated.
```

```python
pair_column = get_max_correlated_column(corr_df['Fuel Information.Highway mpg'])
```

```python
print('Maximum Correlation is: \n ',pair_column)
```
