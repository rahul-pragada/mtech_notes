### Tuples - Methods and Description

* Ordered
* Immutable
* Allows duplicate values

```python
tuple_a = ()
tuple_b = tuple()
tuple_c = tuple(('Earth','Mars','Death Star','Krypton'))
print(tuple_a, tuple_b, tuple_c)
```

```python
type(tuple_c)
```

```python
tuple_d = tuple(('Earth','Mars','Death Star','Krypton','Earth'))
tuple_d
```

```python
print(tuple_d[1:])
print(tuple_d[1:4])
print(tuple_d[:4])
print(tuple_d[:-2])
print(tuple_d[-3:-1])
print(tuple_d[::-1])
```

```python
tuple_fruits = ("apple", "banana", "cherry", "strawberry", "raspberry")

(green, yellow, *red) = tuple_fruits

print(green)
print(yellow)
print(red)
```

### Sets - Methods and Description

* unordered
* unchangeable (immutable)
* does not allow duplicate values

```python
set_a = {}
set_b = set()
set_c = set(('alpha','beta','gamma','delta','epsilon','kappa','sigma','mu'))
```

```python
print(set_a,set_b,set_c)
print(type(set_c))
```

```python
len(set_c)
```

```python
#Adding new element to a set. 
set_d = set_c.copy()
set_d.add('pi')
```

```python
set_e = {'leaves','stems','root','branches'}
set_e.update(set_d) #Any iterable can be used with this update method.
set_e
```

```python
set_e.remove('pi')
set_e
```

```python
set_e.remove('phi')
```

```python
set_e.discard('alpha')
set_e
```

```python
set_e.discard('phi')
set_e
```

```python
set_e.pop()
set_e
```

```python
set_e.clear()
set_e
```

```python
del set_e
```
