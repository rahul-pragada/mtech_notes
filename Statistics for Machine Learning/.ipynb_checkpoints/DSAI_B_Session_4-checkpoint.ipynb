```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
#T shirt example--  for the last 2 years ,Avaerage Defect in the T shirt manufacturing at Allen solly i
#5%
```

```python
#i am going to test 20 sample of tshirt
n=20
p=.05
#What is the probability that 3 tshirts are defective from the sample?
x=3
x=np.arange(20)
pval=stats.binom.pmf(x,n,p)
```

```python
plt.plot(x,pval,"o--")
plt.xticks(x)
plt.show()
```

```python
#What is the probability that maximum of 3 tshirts defective?
stats.binom.pmf(0,n,p)+stats.binom.pmf(1,n,p)+stats.binom.pmf(2,n,p)+stats.binom.pmf(3,n,p)
```

```python
stats.binom.cdf(3,n,p)
```

```python
# central limit theorm
w=np.random.randn(85000)
```

```python
sns.distplot(w)
```

```python
w.max(),w.min()
```

```python
w
```

```python
me=3
sd=0.5
```

```python
w_pop=(w*sd)+me
```

```python
w_pop
```

```python
w_pop.min(),w_pop.max()
```

```python
sns.boxplot(w_pop)
```

```python
# classical method
t=[]
for itr in np.arange(200):
    
    samp=[]
    for i in np.arange(500):
        samp.append(np.random.choice(w_pop))
    t.append(np.mean(samp))
```
