```python
plt.figure(figsize=(15,5))
sns.lineplot(data=l3,marker='o',linestyle='--',color='g')
plt.xlim(0,20)
plt.ylim(0,1)
plt.xlabel('X value')
plt.ylabel('Y Value')
plt.grid(True)
plt.title('Line Graph')
```
