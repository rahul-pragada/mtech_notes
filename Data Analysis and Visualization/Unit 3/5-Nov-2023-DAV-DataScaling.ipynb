unit of measurement

Scaling - Bring something to measurable entity

Data Scaling = Bring data to a scalable level.
Say one feature is running in millions and another in 1000s and another in 10s they bring them to a scale to cross comparison would easier. And it is called Sclaing

Feature scaling

z score = (x - Mean) / sqrt(SD)

Nor ==> 0 to 1

Mean ==> 0,

SD ==> 1

Euclidean Distance used to combine 2 features to a comparable mode

standardization : x-stand = x - mean(x) /Standard Deviation (x)

Normalization : x-norm = (x -min(x)) /(max(x)- min(x))

standar scaler: x-scaled = xi -xmean / sigma

robust scalar: xscaled = xi - xmedian / IQR

Min max scaler:  xscaled = xi - xmean / xmax - xmin

max abs scaler

max abs scaler=  xscaled = (xi - max(|x|)) / max(|x|)

```python
import numpy as np
import pandas as pd
import sklearn as sn
#conda install scikit-learn-intelex
```

```python
df = pd.read_csv("sal.csv")
```

```python
df
```

```python
#Absolute max scaling
#max abs scaler= xscaled = (xi - max(|x|)) / max(|x|)
max_val = np.max(np.abs(df),axis=0)
print(max_val)
type(max_val)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df)
scaled_newtable = pd.DataFrame(scaled_data,columns=df.columns)
scaled_newtable.head(10)
```

```python
from sklearn.preprocessing import Normalizer
scaler = Normalizer()
scaled_data = scaler.fit_transform(df)
norm_sscaled_newtable = pd.DataFrame(scaled_data,columns=df.columns)
norm_sscaled_newtable.head(10)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
std_sscaled_newtable = pd.DataFrame(scaled_data,columns=df.columns)
std_sscaled_newtable.head(10)
```

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
scaled_data = scaler.fit_transform(df)
Robust_std_sscaled_newtable = pd.DataFrame(scaled_data,columns=df.columns)
Robust_std_sscaled_newtable.head(10)
```

```python
wine= pd.read_csv("winequality-red.csv", usecols=["fixed acidity", "volatile acidity", "citric acid", "alcohol"])
```

```python
wine.head()
```

```python
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(wine)
MinMAxscaled_newtable = pd.DataFrame(scaled_data,columns=wine.columns)
MinMAxscaled_newtable.head(10)
```

```python
#from sklearn.preprocessing import Normalizer
#from sklearn import preprocessing ----- this also works. another method
scaler = Normalizer()
norm_scaled_data = scaler.fit_transform(wine)
norm_sscaled_newtable = pd.DataFrame(scaled_data,columns=wine.columns)
norm_sscaled_newtable.head(10)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(wine)
std_sscaled_newtable = pd.DataFrame(scaled_data,columns=wine.columns)
std_sscaled_newtable.head(10)
```

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
scaled_data = scaler.fit_transform(wine)
Robust_std_sscaled_newtable = pd.DataFrame(scaled_data,columns=wine.columns)
Robust_std_sscaled_newtable.head(10)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (15,4)) ##--- Check later output should come horizantally for comparison rather than vertically
plt.subplot(1,5,1)
sns.kdeplot(wine['fixed acidity'],color= 'red')
sns.kdeplot(wine['citric acid'],color= 'green')
sns.kdeplot(wine['volatile acidity'],color= 'blue')
sns.kdeplot(wine['alcohol'],color= 'brown')
plt.title('Without scaling')

plt.subplot(1,5,2)
sns.kdeplot(MinMAxscaled_newtable['fixed acidity'],color= 'red')
sns.kdeplot(MinMAxscaled_newtable['citric acid'],color= 'green')
sns.kdeplot(MinMAxscaled_newtable['volatile acidity'],color= 'blue')
sns.kdeplot(MinMAxscaled_newtable['alcohol'],color= 'brown')
plt.title('MinMAxscaled scaling')

plt.subplot(1,5,3)
sns.kdeplot(norm_sscaled_newtable['fixed acidity'],color= 'red')
sns.kdeplot(norm_sscaled_newtable['citric acid'],color= 'green')
sns.kdeplot(norm_sscaled_newtable['volatile acidity'],color= 'blue')
sns.kdeplot(norm_sscaled_newtable['alcohol'],color= 'brown')
plt.title('norm_sscaled_newtable scaling')

plt.subplot(1,5,4)
sns.kdeplot(std_sscaled_newtable['fixed acidity'],color= 'red')
sns.kdeplot(std_sscaled_newtable['citric acid'],color= 'green')
sns.kdeplot(std_sscaled_newtable['volatile acidity'],color= 'blue')
sns.kdeplot(std_sscaled_newtable['alcohol'],color= 'brown')
plt.title('std_sscaled_newtable scaling')

plt.subplot(1,5,5)
sns.kdeplot(Robust_std_sscaled_newtable['fixed acidity'],color= 'red')
sns.kdeplot(Robust_std_sscaled_newtable['citric acid'],color= 'green')
sns.kdeplot(Robust_std_sscaled_newtable['volatile acidity'],color= 'blue')
sns.kdeplot(Robust_std_sscaled_newtable['alcohol'],color= 'brown')
plt.title('Robust_std_sscaled_newtable scaling')

plt.tight_layout()
plt.show()
```
