```python
fun=lambda x,y:x if x>y else y
print(fun(3,4))
```

```python
# lambda with map
sample_list = [1,2,3,4]
seq=list(map(lambda x: x**2, sample_list))
seq               
```

```python
#lambda with filter
num_list = list(range(14))
seq=list(filter(lambda x:x%3==0, num_list))
seq
```

```python
# Lambda with reduce
from functools import reduce
reduce(lambda a,b:a+b, [3,5,6,7,8,9])
```

```python
# lambda with accumulate
from itertools import accumulate
num_seq=list(range(10))
tuple(accumulate(num_seq, lambda x,y:x+y))
```
